```python
# Importando as bibliotecas do Python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Filtrando Dados

```python
# Lendo o arquivo CSV
df = pd.read_csv("titanic.csv")
```

```python
# Mostrando os primeiros registros da Tabela.
df.head()
```

```python
# Visão geral dos dados
print(df.info())
```

```python
# Limpeza de dados
# Preenchimento de valores ausentes
df["Embarked"].fillna("S", inplace=True)
df["Age"].fillna(df["Age"].mean(), inplace=True)
```

```python
# Conversão de variáveis
df["Sex"] = df["Sex"].astype("category")
df["Embarked"] = df["Embarked"].astype("category")
```

```python
# Criação de novas variáveis
df["Idade_Grupo"] = pd.cut(df["Age"], [0, 18, 30, 45, 60, 80], labels=["Criança", "Jovem", "Adulto", "Meia-Idade", "Idoso"])
```

# Mais Slicing de Dados com DataSet

```python
# Distribuição da variável "Survived"
sns.countplot(x="Survived", data=df)
plt.show()
```

```python
# Distribuição da variável "Idade" por sexo
sns.displot(x="Age", hue="Sex", data=df, kind="kde")
plt.show()
```
